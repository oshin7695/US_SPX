```python
def exit_strategy(df):
    filtered_df_list = []
    
    # Group by each date
    for date, group in df.groupby('date'):

        # Step 1: Get the first 'call_pnl' value for the date (first_straddle)
        first_call = group['call_price'].iloc[0]
        
        # Step 2: Define the stop-loss level
        stop_loss_level = 1.4 * first_call
        
        # Step 3: Apply the stop-loss logic
        stop_loss_triggered = False  # Track if the stop-loss condition has been hit
        
        for i in range(len(group)):
            if not stop_loss_triggered and group['call_price'].iloc[i] >= stop_loss_level:
                stop_loss_triggered = True
                a = i
            
            # Forward-fill the stop-loss level if triggered
            if stop_loss_triggered:
                group.loc[group.index[i], 'call_price'] = group.loc[group.index[a], 'call_price']
                group['call_stop_time'] = group.loc[group.index[a], 'time']
        
        # Step 4: Add the adjusted group to the filtered list
        filtered_df_list.append(group)
    
    # Combine all the filtered groups
    return pd.concat(filtered_df_list, ignore_index=True)
```

```python
filtered_dfs = []

for date, group in df4.groupby('date'):
    first_stop = group['first_stop'].iloc[0]

    if first_stop == 'call':
        condition = (group['time'] > group['call_stop_time'].iloc[0]) & (group['call_price1'] < group['call932'].iloc[0])
    elif first_stop == 'put':
        condition = (group['time'] > group['put_stop_time'].iloc[0]) & (group['put_price1'] < group['put932'].iloc[0])
    else:
        continue  # Skip if first_stop is neither 'call' nor 'put'
    
    # Find the first index where the condition is met
    valid_indices = group.index[condition]
    
    if not valid_indices.empty:
        first_valid_index = valid_indices[0]
        filtered_dfs.append(group.loc[first_valid_index:])  # Select from first valid index onwards

df_filtered = pd.concat(filtered_dfs) if filtered_dfs else pd.DataFrame()
```

```python
def exit_strategy_rentry(df):
    filtered_df_list = []
    
    # Group by each date
    for date, group in df.groupby('date'):

        # Step 1: Get the first 'put_pnl' value for the date (first_straddle)
        first_put = group['rentry_price'].iloc[0]
        
        # Step 2: Define the stop-loss level
        stop_loss_level = 1.5 * first_put
        
        # Step 3: Apply the stop-loss logic
        stop_loss_triggered = False  # Track if the stop-loss condition has been hit
        
        for i in range(len(group)):
            if not stop_loss_triggered and group['rentry_price'].iloc[i] >= stop_loss_level:
                stop_loss_triggered = True
                a = i
            
            # Forward-fill the stop-loss level if triggered
            if stop_loss_triggered:
                group.loc[group.index[i], 'rentry_price'] = group.loc[group.index[a], 'rentry_price']
                group['reentry_stop_time'] = group.loc[group.index[a], 'time']
        
        # Step 4: Add the adjusted group to the filtered list
        filtered_df_list.append(group)
    
    # Combine all the filtered groups
    return pd.concat(filtered_df_list, ignore_index=True)
```
